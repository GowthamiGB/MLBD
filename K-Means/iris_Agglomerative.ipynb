```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
iris = datasets.load_iris()
iris
```

```python
X = iris.data[:,:2]
X
```

```python
from scipy.cluster.hierarchy import dendrogram , linkage
import matplotlib.pyplot as plt
linkage_data=linkage(X,method='ward',metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
x,y=zip(*X)

hierarchical_cluster=AgglomerativeClustering(n_clusters=3, metric='euclidean',linkage='ward')
labels=hierarchical_cluster.fit_predict(X)
plt.scatter(x,y,c=labels)
plt.show()
```

```python
plt.scatter(X[labels==0,0],X[labels==0,1],s=100,c='red',label='iris1')
plt.scatter(X[labels==1,0],X[labels==1,1],s=100,c='blue',label='iris2')
plt.scatter(X[labels==2,0],X[labels==2,1],s=100,c='pink',label='iris3')

# plt.scatter(hierarchical_cluster.cluster_centers_[:,0],hierarchical_cluster.cluster_centers_[:,1],s=300,c='Yellow',label='Centroids')
plt.title('Iris dataset')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.legend()
plt.show()
```
