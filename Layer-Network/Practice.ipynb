```python
x1=[0,0,1,1]
x2=[0,1,0,1]
w=[1,1]
```

AND_GATE

```python
and_gate=[]
for i in range(len(x1)):
  and_gate.append(x1[i]*w[0] + x2[i]*w[1])
print("x1  x2  yin  y")

for i in range(len(and_gate)):
  print(f'{x1[i]}   {x2[i]}    {and_gate[i]}   {1 if and_gate[i]>=max(and_gate) else 0}')#The f before a string, also known as an f-string in Python,
  #is used to create formatted strings. It's a way to embed expressions inside string literals, allowing for easy string formatting with variables and expressions.
```

OR_GATE

```python
or_gate=[]
for i in range(len(x1)):
  or_gate.append(x1[i]*w[0]+x2[i]*w[1])
print('x1  x2  yin  y')

for i in range(len(or_gate)):
  print(f'{x1[i]}   {x2[i]}   {or_gate[i]}    {1 if or_gate[i]>min(or_gate) else 0}')
#print(f'{x1[i]}    {x2[i]}    {orG[i]}     {0 if x1[i] == 0 and x2[i] == 0 else 1}') can give directly also
```

XOR_GATE

```python
x1 = [0, 0, 1, 1]
x2 = [0, 1, 0, 1]
w = [1, -1, -1, 1]

xor_gate1 = []
xor_gate2 = []

print("X1   X2   y1in  y1  y2in  y2  y")
for i in range(len(x1)):
    xor_gate1.append(x1[i] * w[0] + x2[i] * w[2])
    xor_gate2.append(x2[i] * w[3] + x1[i] * w[1])

    y1 = 1 if xor_gate1[i] >= 1 else 0
    y2 = 1 if xor_gate2[i] >= 1 else 0

    y = y1 + y2

    print(f'{x1[i]}    {x2[i]}    {xor_gate1[i]:<5} {y1:<3} {xor_gate2[i]:<5} {y2:<3} {y:<3}')

    #The <5 specifies left alignment with a width of 5 characters for the y1in and y2in columns,
    #and <3 specifies left alignment with a width of 3 characters for the y1, y2, and y columns.
```
