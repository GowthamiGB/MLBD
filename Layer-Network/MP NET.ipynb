```python
# x1=[0,0,1,1]
# x2=[0,1,0,1]
# w=[1,1]
x1 = [int(x) for x in input("Enter values for x1 (separated by space): ").split()] #list comprehension
x2 = [int(x) for x in input("Enter values for x2 (separated by space): ").split()]
w = [int(w) for w in input("Enter values for weights w1 and w2 (separated by space): ").split()]
```

AND_GATE

```python
and_gate=[]
for i in range(len(x1)):
  and_gate.append(x1[i]*w[0] + x2[i]*w[1])
print("x1  x2  yin  y")

for i in range(len(and_gate)):
  print(f'{x1[i]}   {x2[i]}    {and_gate[i]}   {1 if and_gate[i]>=max(and_gate) else 0}')#The f before a string, also known as an f-string in Python,
  #is used to create formatted strings. It's a way to embed expressions inside string literals, allowing for easy string formatting with variables and expressions.
```
