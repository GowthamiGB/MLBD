```python

def y_Input(x1,x2,w1,w2,b):
    yin = (x1*w1) + (x2*w2) + b
    return yin
```

```python
def activation_function(inp,theta):
    if inp >= theta:
        return 1
    else:
        return -1
```

```python
def weightsCalculation(X1,X2,t):
    w1 = w2 = b = 0
    alpha = theta = 1
    for dump in range(len(X1)):
        target = t[dump]
        y = activation_function(y_Input(X1[dump],X2[dump],w1,w2,b),theta)
        while y != target:
            w1 = w1 + (alpha*target*(X1[dump]))
            w2 = w2 + (alpha*target*(X2[dump]))
            b = b + (alpha*target)
            y = activation_function(y_Input(X1[dump],X2[dump],w1,w2,b),theta)
    print(f"Final Weights : W1 = {w1}, W2 = {w2}, Bias = {b}")
```

# AND GATE

```python
X1 = [1,1,-1,-1]
X2 = [1,-1,1,-1]
t =  [1,-1,-1,-1]
weightsCalculation(X1,X2,t)
```

# OR GATE

```python
X1 = [1,1,-1,-1]
X2 = [1,-1,1,-1]
t =  [1,1,1,-1]
weightsCalculation(X1,X2,t)
```

# NOT GATE

```python
X1 = [1,-1]
X2 = [1,-1]
t =  [-1,1]
weightsCalculation(X1,X2,t)
```

# NOR GATE

```python
X1 = [1,1,-1,-1]
X2 = [1,-1,1,-1]
t =  [-1,-1,-1,1]
weightsCalculation(X1,X2,t)
```

# NAND GATE

```python
X1 = [1,1,-1,-1]
X2 = [1,-1,1,-1]
t =  [-1,1,1,1]
weightsCalculation(X1,X2,t)
```
