```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
import os
import cv2
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data=tf.keras.utils.image_dataset_from_directory('/content/drive/MyDrive/MLB-LAB/K-Means/data')
data
```

```python
data.class_names
```

```python
data=data.map(lambda x,y:(x/255,y))
```

```python
scaled_iterator=data.as_numpy_iterator()
```

```python
scaled_batch=scaled_iterator.next()
```

```python
model=Sequential()
```

```python
model.add(Conv2D(16,(3,3),1, activation='relu', input_shape=(256,256,3)))
```

```python
model.add(MaxPooling2D())
```

```python
model.add(Conv2D(32,(3,3),1, activation='relu'))
```

```python
model.add(MaxPooling2D())
```

```python
model.add(Conv2D(16,(3,3),1, activation='relu'))
```

```python
model.add(MaxPooling2D())
```

```python
model.add(Flatten())
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(data,epochs=10)
```

```python
model.summary()
```

```python
img=cv2.imread("/content/drive/MyDrive/MLB-LAB/K-Means/data/cat/kitty-cat-kitten-pet-45201.jpeg")
img
```

```python
resize=tf.image.resize(img,(256,256))
```

```python
predictedClass=model.predict(np.expand_dims(resize/255,0))
```
