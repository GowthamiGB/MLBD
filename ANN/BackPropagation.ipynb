```python
dataset =pd.read_csv("drive/My Drive/MLB-LAB/Social_Network_Ads.csv") #importing the dataset(step-3)
data=dataset
data
```

```python
X=data.iloc[:,1:4].values #independent variale-X ->if .values is removed proper two columns are printed.But with .values it prints as an array.
X
```

```python
Y=data.iloc[:,4].values
Y
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

encoder = LabelEncoder()
X[:,0] = encoder.fit_transform(X[:,0]) #for gender
X
```

```python
X=X[:,0:]
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#As we are using large number of inputs-> to avoid scaling difference or to take data to same scale we use standard scalar
#Generally if we have attributes with one value and other with varying value it helps to scale to a single value which will be in same range.
from sklearn.preprocessing import StandardScaler #when numbers are in two  different levels-we use Standard Scaler to normalize data here age and salary
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_test
```

**NOW LET'S MAKE ANN**

```python
import keras #for deep nural networks
from keras.models import Sequential #feed forward network
from keras.layers import Dense #fully-connected->each and every node is connected to all the nodes in next layer
```

```python
#Initializing the ANN
classifier=Sequential()
```

```python
#Adding the input layer and first hidden layer->
# input layer has 11 input so to decide the nodes in hidden layer->1.hyperparameter tuning 2.units=[3(input)+1(output)/2]
#kernel_initializer-> is the weights- here uniform is given (same random number is assigned)
# activation functions-> 1. relu=rectified linear unit used for all the hidden layer-> gives output either 0 or 1
# 2. sigmoidal-> used for all the output layer->gives probability of values between 0 and 1->either bipolar or binary
# input_dim-> input values
classifier.add(Dense(units=2, kernel_initializer='uniform',activation='relu',input_dim=3))
```

```python
# Adding second hidden layer
classifier.add(Dense(units=2, kernel_initializer='uniform',activation='relu')) #no need to give input again as dense function will take care of the connection
```

```python
# Add the output layer->for multiple outputs use activation function as softmax
classifier.add(Dense(units=1, kernel_initializer='uniform',activation='sigmoid'))
```

```python
# Compiling the ANN - adam optimizes weights using Stochastic Gradient Descent -> to reduce error gradually
# loss function for binary classification is binary_crossentropy -> for 1 output
# loss function for multi-class classification model where there are two or more output labels category_crossentropy
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Fitting the ANN to the training set
classifier.fit(X_train,Y_train,batch_size=10,epochs=10)
#Total input=10,000 Training=8,000 ,weight updation is for every batch (how many data points are privided), epoch-> this 8,000 how many times it is repeated
```

```python
Y_pred=classifier.predict(X_test)
Y_pred
```

```python
# As output is sigmoidal -> now predicting the test set results
Y_pred=classifier.predict(X_test)
Y_pred=Y_pred>0.45 #greater than 0.45 is true else false
Y_pred
```

```python
Y_test
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
cm
```

```python
from sklearn import metrics
prediction=metrics.accuracy_score(Y_test,Y_pred)
print("accuracy:",prediction *100,"%")
```
